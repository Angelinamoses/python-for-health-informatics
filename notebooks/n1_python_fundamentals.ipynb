## Type Conversion: Converting Data into the Correct Format

Data collected from users, medical devices, hospital systems, and Electronic Health Records (EHRs) does not always arrive in the format we need.

Computers treat different types of data differently. A patient's age should be treated as a number, while a patient's name should be treated as text.

When data is stored in an inappropriate format:

* Calculations may fail
* Results may become inaccurate
* Clinical dashboards may display incorrect information
* Machine learning models may produce unreliable predictions

Type conversion is the process of changing one data type into another.

| Function | Purpose |
|----------|---------|
| int() | Convert to integer |
| float() | Convert to decimal number |
| str() | Convert to text |
| bool() | Convert to Boolean value |

### Why Type Conversion Matters in Healthcare

Consider a patient registration form.

Healthcare applications constantly perform calculations involving:

### Common Type Conversion Errors

One of the most common beginner mistakes is assuming that numbers stored as text behave like numbers.

```python
age1 = "25"
age2 = "5"

print(age1 + age2)
```

Output:

Python joins the text together instead of performing addition.

This operation is called string concatenation.

```python
#example:
age1 = int("25")
age2 = int("5")

print(age1 + age2)
```

Interpretation

Output:

30

After conversion, Python treats both values as numbers and performs arithmetic.

## Healthcare Example: BMI Calculation

Body Mass Index (BMI) is commonly used to assess whether a person's weight is appropriate for their height.

The formula is:

BMI = Weight (kg) ÷ Height² (m²)

Although BMI is not a diagnostic measure, it is frequently used in healthcare research and population health studies.

```python
weight = 72
height = 1.65

bmi = weight / (height ** 2)

print("BMI:", round(bmi, 2))
```

## Interpretation

Python first calculates the square of the height.

height² = 1.65 × 1.65

It then divides the weight by the squared height.

The resulting BMI can be used for screening and population-level analysis.

# Logical Operators

Logical operators combine multiple conditions.

They help healthcare systems make decisions based on several pieces of information at the same time.

## Logical Operators

| Operator | Meaning |
|----------|---------|
| and | Both conditions must be true |
| or | At least one condition must be true |
| not | Reverses a condition |

```python
has_fever = True
has_cough = True

print(has_fever and has_cough)
```

## Interpretation

Output:

True

The result is True because both conditions are True.

The `and` operator requires every condition to be satisfied.

```python
has_fever = True
has_cough = False

print(has_fever or has_cough)
```

## Interpretation

Output:

True

The `or` operator returns True when at least one condition is True.

This type of logic is commonly used in clinical decision support systems.

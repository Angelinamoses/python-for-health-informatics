# What control flow means
## Goal: 
  Understand how Python decides what to run, when to run, and how long to run it

### Topics:
* if, elif, else
* nested conditions
* match-case
* for loops
* while loops
* break
* continue
* pass

### 1. if_else
Runs code based on condition.

if condition:
#### block runs if condition is True
else:
#### block runs if condition False

```python
#ex:
age = 22
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# elif: blood pressure classification:

bp = 165

if bp < 120:
    print("Normal")

elif bp < 140:
    print("Elevated")

elif bp < 180:
    print("High")

else:
    print("Critical")
```

### Health Informatics relevance:
* Checking if BMI is in a risk range
* Validating patient age eligibility
* Flagging abnormal lab values

### 2. nested_conditions
Concept:
 An if inside another if.

age = 45
bp = 150

if age > 40:
    if bp > 140:
        print("High cardiovascular risk")
    else:
        print("Monitor blood pressure")
else:
    print("Lower age-related risk")

### Use in Health Informatics:
* Decision support systems
* Clinical triage logic
* Multifactor risk assessment

### 3. match_case
Concept: A Cleaner way to handle multiple fixed choices.

```python
day = "Monday"

match day:
    case "Monday":
        print("Start of the week")
    case "Friday":
        print("Almost weekend")
    case _:
        print("Regular day")
```

### Health Informatics relevance
* Mapping disease categories
* Handling status codes
* Categorizing lab test types

### 4. for_loops
Concept: Used when you know how many items you wnt to iterate through.

```python
patients = ["Asha", "Ravi", "Meena"]

for patient in patients:
    print(patient)
```

```python
for i in range(6):
    print(i)
```

### Use in health informatics
* Loop through patient records
* Process test results
* Generate summaries from datasets

### 5.while_loops
Concept: Keeps running while a condition remains True.

```python
count = 0

while count <5:
    print(count)
    count +=1
```

### Use in Health informatics
* Repeat until valid input is received
* Keep polling until data arrives
* Retry failed operations

### 6.break
Concept: Stops a loop completely

```python
for num in range(10):
    if num == 5:
        break
    print(num)
```

### Meaning:
Loop stops when num==5.

### Health informatics use
* Stop scanning once a target record is found
* Stop when an alert threshold is reached

### 7.continue
Concept: Skips the current loop iteration and moves to the next one.

```python
for num in range(5):
    if num == 2:
        continue
    print(num)
```

### Meaning:
2 is skipped.
### Health Informatics use
* Ignore invalid rows
* Skip missing values
* Skip records that fail a rule

### 8.pass
Concept: Does nothimg. Literally a placeholder.

```python
for i in range(5):
    pass
```

### When to use:
* While planning code structure
* When a block is required syntactically but logic is not ready yet

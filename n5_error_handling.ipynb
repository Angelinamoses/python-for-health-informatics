### try_except

#### Handling exceptions with Try and Except

#### Why Try_Except?

```python
# Without error handling:

age = int("abc") # program will crash here with a ValueError because "abc" cannot be converted to an integer.
```

```python
# with error handling:

try:
    age = int("abc")  # This will raise a ValueError, but it will be caught by the except block.
except:
    print("Invalid input. Please enter a valid integer.")
```

```python
# example

try:
    glucose_level = int(input("Enter the patient's glucose level: "))

except:
    print("Please enter numbers only")
```

#### Catch Specific Errors

Bad -

except:

better -

except ValueError:

```python
# example:

try:
    age = int(input())

except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
# Multiple Exceptions

try:
    result = 10 / int(input())
except ValueError:
    print("Invalid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Else block: Runs if no exceptions were raised in the try block.

try:
    age = int(input())
except ValueError:
    print("Invalid")
else:
    print("Valid age")
```

```python
# Finally block: Runs no matter what, even if an exception was raised or not.

try:
    file = open("patients.txt")
except FileNotFoundError:
    print("File missing")
finally:
    print("Operation complete")
```

```python
# Healthcare Example: Patient glucose entry

try:
    glucose = int(input("Enter glucose value: "))
except ValueError:
    print("Glucose must be numeric")
else:
    print("Data saved")
```

### raise.ipynb

#### Raising Exceptions: Creating Errors Intentionally

Sometimes python doesn't know something is wrong, but You do.

Use: 

raise

* to create your own exception.

```python
# example:

age = -5
```

```python
# instead of allowing the program to continue with an invalid age, we can raise an exception to indicate that something went wrong.

age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

### Why Raise Exists

python sees:  age = -5

and thinks, it's fine. But we know Negative age makes no sense


```python
# Healthcare example: Patient temperature

temp = -100
```

```python
# -100 is not a valid body temperature, so we can raise an exception to indicate that the value is out of range.

if temp < 0:
    raise ValueError(
        "Temperature cannot be negative"
        )
```

```python
# example: Validate BMI

def calculate_bmi(weight, height):

    if weight <= 0:
        raise ValueError("Weight must be greater than zero")
    if height <= 0:
        raise ValueError("Height must be greater than zero")
    
    bmi = weight / (height ** 2)
    return bmi
```

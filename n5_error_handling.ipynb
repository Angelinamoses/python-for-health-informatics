### try_except

#### Handling exceptions with Try and Except

#### Why Try_Except?

```python
# Without error handling:

age = int("abc") # program will crash here with a ValueError because "abc" cannot be converted to an integer.
```

```python
# with error handling:

try:
    age = int("abc")  # This will raise a ValueError, but it will be caught by the except block.
except:
    print("Invalid input. Please enter a valid integer.")
```

```python
# example with CI/CD support

try:
    if IS_CI:
        glucose_level = int("abc")  # Default invalid input for CI/CD to trigger exception
    else:
        glucose_level = int(input("Enter the patient's glucose level: "))
except:
    print("Please enter numbers only")
```

#### Catch Specific Errors

Bad -

except:

better -

except ValueError:

```python
# example with CI/CD support:

try:
    if IS_CI:
        age = int("abc")  # Default invalid input for CI/CD
    else:
        age = int(input())
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
# Multiple Exceptions with CI/CD support

try:
    if IS_CI:
        result = 10 / int("0")  # Default zero for CI/CD to trigger ZeroDivisionError
    else:
        result = 10 / int(input())
except ValueError:
    print("Invalid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Else block: Runs if no exceptions were raised in the try block.

try:
    if IS_CI:
        age = int("25")  # Default valid input for CI/CD
    else:
        age = int(input())
except ValueError:
    print("Invalid")
else:
    print("Valid age")
```

```python
# Finally block: Runs no matter what, even if an exception was raised or not.

try:
    file = open("patients.txt")
except FileNotFoundError:
    print("File missing")
finally:
    print("Operation complete")
```

```python
# Healthcare Example: Patient glucose entry with CI/CD support

try:
    if IS_CI:
        glucose = int("abc")  # Default invalid input for CI/CD
    else:
        glucose = int(input("Enter glucose level: "))
except ValueError:
    print("Glucose must be numeric")
```

### constructors

### Constructors: Initializing Objects Automatically

Why Constructors?

```python
# without constructor:

class Patient:

    pass
```

```python
patient = Patient()

patient.name = "Asha"
patient.age = 24   # must set everything manually
```

```python
# with constructor:
class Patient:

    def __init__(self, name, age):   #__init__ is the constructor method, runs automatically when an instance is created
        self.name = name
        self.age = age
```

```python
# Health informatics example:

class LabResult:

    def __init__(self, test_name, result_value):
        self.test_name = test_name
        self.result_value = result_value
```

```python
result_value = LabResult("Blood Glucose", 5.6)

print(result_value.test_name)  # Output: Blood Glucose
print(result_value.result_value)  # Output: 5.6
```

### inheritance

### Inheritance: Reusing Existing Code

Why Inheritance?
_Doctor and Patient share common properties, instead of rewriting it_

Reuse it.

```python
class Person:

    def greet(self):
        print("Hello!")
```

```python
class Patient(Person):  # Patient class inherits from Person class
    pass
```

```python
patient = Patient()

patient.greet()  # Output: Hello!
```

```python
class Doctor(Person):  # Doctor class also inherits from Person class
    
    def prescribe(self):
        print("Medication prescribed.")
```

```python
doctor = Doctor()

doctor.greet()  # Output: Hello!
doctor.prescribe()  # Output: Medication prescribed.
```

### polymorphism

### Polymorphism: One Interface, Many Forms

What is Polymorphism?
_same method, but different behaviour_

```python
class Doctor:

    def role(self):
        print("Doctor treating a patient.")

class Nurse: 
    def role(self):
        print("Nurse assisting a patient.")

class Patient:

    def role(self):
        print("Patient receiving care.")
```

```python
doctor = Doctor()
nurse = Nurse()
patient = Patient()

doctor.role()  # Output: Doctor treating a patient.
nurse.role()   # Output: Nurse assisting a patient.
patient.role()  # Output: Patient receiving care.
```

_notice:_

role()

was called every time.

same method name.

different behaviour.

This is polymorphism.

### encapsulation

### Encapsulation: Protecting Data

Why Encapsulation?

_Not everyone should modify sensitive healthcare data._

```python
class Patient:

    def __init__(self, age):
        self.__age = age  # This is a private attribute, not accessible outside the class
```

```python
patient = Patient(30)
print(patient._Patient__age)  # This will raise an AttributeError since __age is private
```

```python
# getter method:

class Patient:

    def __init__(self, age):
        self.__age = age  # This is a private attribute, not accessible outside the class

    def get_age(self):
        return self.__age
```

```python
patient = Patient(30)
print(patient.get_age())  # Output: 30
```

```python
# setter method:

class Patient:

    def __init__(self, age):
        self.__age = age # This is a private attribute, not accessible outside the class

    def set_age(self, age):

        if age > 0:
            self.__age = age
        
```

### magic_methods

### Magic methods: Making Objects Behave Like Built-in Types

#### What are magic methods?

_special methods automatically called by Python._

```python
# str():

class Patient:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Patient: {self.name}"
```

```python
patient = Patient(30)

print(patient)  # Output: Patient: 30
```

#### Why str Matters

_used for:_

* reports
* dashboards
* patient summaries
* user interfaces

```python
# Health informatics example:

class LabResult:

    def __init__(self, test_name, result_value):
        self.test_name = test_name
        self.result_value = result_value

    def __str__(self):
        return f"LabResult: {self.test_name} = {self.result_value}"
```

```python
result_value = LabResult("Blood Glucose", 5.6)

print(result_value)  # Output: LabResult: Blood Glucose = 5.6
```

```python
# repr(): Developer friendly string representation of an object, often used for debugging

class Patient:

    def __init__(self, patient_id, age):
        self.patient_id = patient_id
        self.age = age

    def __repr__(self):
        return f"Patient('{self.patient_id}' , {self.age})"

```

```python
patient = Patient("P001", 34)

print(repr(patient))
```

```python
# len()

class PatientList:

    def __init__(self):
        self.patients = [
            "P001",
            "P002",
            "P003"
        ]

    def __len__(self):
        return len(self.patients)
```

```python
records = PatientList()

print(len(records))
```

```python
# add()

class Cost:

    def __init__(self, amount):
        self.amount = amount

    def __add__(self, other):
        return self.amount + other.amount
```

```python
visit1 = Cost(1000)
visit2 = Cost(850)

print(visit1 + visit2)
```

```python
# getitem()

class Vitals:

    def __init__(self):
        self.readings = [120, 125, 118]

    def __getitem__(self, index):
        return self.readings[index]
```

```python
vitals = Vitals()

print(vitals[0])
```

```python
# contains()

class Diagnoses:

    def __init__(self):
        self.codes = [
            "I10",
            "E11",
            "J45"
        ]

    def __contains__(self, code):
        return code in self.codes
```

```python
diagnoses = Diagnoses()

print("J45" in diagnoses)
```

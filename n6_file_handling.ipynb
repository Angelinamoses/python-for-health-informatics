### reading_files

#### Reading Files in Python: Axxessing Stored Data

### What is file reading?

_file reading means opening a file and getting its content into python_

#### A file may contain:

* patient notes
* lab reports
* CSV data
* logs
* exported reports

_python can only process file data after it is read into memory_

### Why reding files matters?
_in health informatics, data is often stored in:_

* .txt
* .csv
* .json
* .log
* exported system files

#### examples:
* reading lab test records
* loading patient lists
* analyzing appointment logs
* reading discharge appointment logs


```python
# Opening a file

file = open("clinic_visits.csv","r")
```

* "clinic_visits.csv" is the file name
* "r" means read mode
* open() gives a file object

_if the file does not exist, Python raises an error_

```python
with open("clinic_visits.csv","r") as file:
    print(file.readline())  # header
    print(file.readline())  # first line of data
```

```python
# read the whole file

file = open("clinic_visits.csv","r")
content = file.read()
print(content)
file.close()
```

* reads the entire files as one big string
* useful for small files
* not ideal for large clinical datasets

```python
# Read one line at a time

file = open("clinic_visits.csv","r")
line1 = file.readline()
line2 = file.readline()

print("header:", line1)
print("first line of data:", line2)

file.close()
```

```python
# Read all lines as a list

file = open("clinic_visits.csv","r")
lines = file.readlines()
print(lines)
file.close()
```

```python
# loop through the file line by line

file = open("clinic_visits.csv","r")

for line in file:
    print(line)

file.close()
```

```python
# split header and rows

file = open("clinic_visits.csv","r")
lines = file.readlines()

header = lines[0].strip()
print("header:", header)

first_row = lines[1].strip()
print("first row:", first_row)
```

```python
# split csv row into values

row = lines[1].strip().split(",")
print("row values:", row)
```

### writing_files

#### Writing Files in Python: Saving Clinic Data and Reports

### What is file writing?

Writing means saving data is from Python into a file.

_In healthcare, this is useful when you want to save:_

* report summaries
* cleaned patient data
* analysis output
* audit notes
* processed records

```python
# write mode

file = open("clinic_report.txt","w")
```

* "w" means write mode
* if the file already exists, it gets overwritten
if it does not exist, python creates it

_So yes, one small typo can erase your file_

```python
# writing a simple text file

file = open("clinic_report.txt","w")
file.write("Clinic Visits Report\n")
file.write("Generated from clinic_visits.csv\n")

file.close()
```

```python
# write a report from the csv file

file = open("clinic_visits.csv","r")
lines = file.readlines()
file.close()

num_records  = len(lines) - 1  # subtract header

report = open("clinic_report.txt", "w")
report.write("Clinic Visits Report\n")
report.write("====================\n")
report.write(f"Total number of visits:{num_records}\n")
report.write(f"Generated from clinic_visits.csv\n")
report.close()
```

```python
# write selected rows into another file: Suppose we only want to save only admitted patients

with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()


with open("clinic_report.txt", "w") as report:
    report.write("Admitted Patients\n")
    report.write("=================\n")
    for line in lines[1:]:  # Skip header
        row = line.strip().split(",")
        admitted = row[9]

        if admitted == "Yes":
            report.write(line)
```

```python
# Write formatted patient information

patient_id = "P001"
patient_name = "John Doe"
department = "General Medicine"

with open("clinic_report.txt", "w") as report:
    report.write(f"Patient ID: {patient_id}\n")
    report.write(f"Patient Name: {patient_name}\n")
    report.write(f"Department: {department}\n")
```

### append_mode

#### Append Mode in Python: Adding New clinical records without erasing old ones

_Append mode adds new content to the end of a file_

"w" overwrites existing content
"a" keeps odd content and adds new content

```python
#append mode
file = open("clinic_report.txt", "a")
```

```python
# append one line

with open("clinic_report.txt", "a") as report:
    report.write("This line is appended to the report.\n")
```

```python
# append a record from the csv file
with open("clinic_visits.csv", "r") as report:
    lines = report.readlines()

new_log = lines[1].strip()   # first patient record

with open("clinic_report.txt", "a") as report:
    report.write(new_log + "\n")
```

```python
# append multiple records from the csv file
with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()

with open("clinic_report.txt", "a") as report:
    for line in lines[1:6]:  # append first 5 patient records
        row = line.strip().split(",")
        admitted = row[9] # check if the patient was admitted
        if admitted == "Yes":
            report.write(line)
```

```python
# append a custom note

with open("clinic_report.txt", "a") as file:
    file.write("2026-06-02: Dataset reviewed for admission trend.\n")
```

### with_context_manager

### Using 'with' in Python: Safer file Handling for Clinic Data

```python
# When you manually open a file, you must close it yourself.

file = open("clinic_visits.csv", "r")
# do some processing with the file 

content = file.read() # read the entire file content into a variable
file.close() # close the file after processing.
# If an error occurs before file.close(), the file may remain open, which can lead to resource leaks. Using 'with' statement is recommended to ensure proper file handling.
```

```python
# append safely using 'with' statement

with open("clinic_report.txt", "a") as file:
    file.write("This is a safe way to append to the file.\n")
```

### why this is the best practice

_using 'with' is cleaner because:_

* it closes files automatically
* it reduces errors
* it makes code easier to read
* it is the standard python style

### csv_manual_python

### Reading and Writing CSV files Manually in Python

_CSV means Comma-Separated Values._

Usually a csv file looks like this:

visit_id,patient_id,age,gender,department,diagnosis,visit_type,wait_time_min,cost_usd,admitted
1001,P001,34,Female,General Medicine,Hypertension,New,18,80,No
1002,P002,67,Male,Cardiology,Diabetes Follow-up,Follow-up,42,220,Yes
...

_CSV is used because it is simple and works well for sharing tabular data._

```python
# Read all lines from the file

with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()

print(lines[0]) # print the header line
print(lines[1]) # print the first data line
```

```python
# separate header and data lines

with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()

header = lines[0].strip().split(",")
print("Header:", header)
```

```python
# convert each row into a list

for line in lines[1:]:
    row = line.strip().split(",")
    print("Data Row:", row)
```

```python
# Converting rows into dictionaries

patients = []

with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()

header = lines[0].strip().split(",")

for line in lines[1:]:
    values = line.strip().split(",")
    record = {}

    for i in range(len(header)):
        record[header[i]] = values[i]
    patients.append(record)

print(patients[0]) # print the first patient record as a dictionary
```

```python
# convert important columns to integers

patients = []
with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()

header = lines[0].strip().split(",")

for line in lines[1:]:
    values = line.strip().split(",")
    record = {}

    for i in range(len(header)):
        record[header[i]] = values[i]

    # Convert age and visit duration to integers
    record["age"] = int(record["age"])
    record["wait_time_min"] = int(record["wait_time_min"])
    record["cost_usd"] = int(record["cost_usd"])

    patients.append(record)

print(patients[0]) # print the first patient record as a dictionary
```

```python
# find admitted patients manually

with open("clinic_visits.csv", "r") as file:
    lines = file.readlines()

for line in lines[1:]:
    row = line.strip().split(",")
    admitted = row[9]

    if admitted == "Yes":
        print("Admitted patient:", row)
```

```python
# write a csv file manually

with open("clinic_output.csv", "w") as file:
    file.write("visit_id,patient_id,age,gender,department,diagnosis,visit_type,wait_time_min,cost_usd,admitted\n")
    file.write("2001,P100,40,Female,Cardiology,Checkup,Follow-up,20,150,No\n")
    file.write("2002,P101,55,Male,Neurology,Migraine,New,35,180,Yes\n")
```

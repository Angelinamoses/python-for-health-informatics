### functions

#### Python functions basics: Reusable Logic for Health Data Tasks

A function is a reusable block of code

```python
# Instead of writing:

print("Patient checked")
print("Vitals recorded")
print("Medication administered")

# We can write: You define it once!!
def check_patient():
    print("Patient checked")
    print("Vitals recorded")
    print("Medication administered")
```

```python
# Then reuse how many times you want:

check_patient()
```

### Why functions matter?
Without functions:

* Code becomes repititive
* debugginh becomes painful
* projects becomes disasters

With functions:

* cleaner logic
* reusable workflows
* esier testing
* modular design

### Real Healthcare EXample

Imagine:

* calculating bmi
* checking diabetes risk
* formatting patient reports
* processing lab results

```python
# Let's check bmi in traditional way:

weight = 70
height = 1.75

bmi = weight / height ** 2
print(bmi)
```

```python
# With functions:

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi
```

```python
# now we can reuse the function with different values:

patient1_bmi = calculate_bmi(70, 1.75)
patient2_bmi = calculate_bmi(80, 1.80)

print("Patient 1 BMI:", patient1_bmi)
print("Patient 2 BMI:", patient2_bmi)
```

### parameters_and_arguments

#### Parameters and Arguments: Making functions dynamics

Core Idea: Functions become powerful when they can accept data.

```python
# instead of:

def greet():
    print("Hello, welcome to the clinic!")
```

```python
# use:

def greet(name):
    print(f"Hello {name}, welcome to the clinic!") # now this function is more flexible and can greet different people by name!
```

```python
#Example usage:
def greet(name): # parameter
    print(name)

greet("John") # argument
```

```python
# multiple parameters:

def patient_info(name, age):
    print(name)
    print(age)
```

```python
patient_info("John", 30)  #positional arguments
```

### return_values

#### Return Values: Sending Data back from Functions

Key Concept: Function can produce results

```python
# Without return:

def add():
    print(2+3)
```

```python
# with return:
def add():
    return 2+3
```

### Why Return matters?

print() shows output

return gives output back to program.

huge difference

```python
# Healthcare Example: BMI Calculator

def bmi(weight, height):
    return weight / (height ** 2)
```

```python
# Usage:
value = bmi(70, 1.75)
print("BMI:", value)
```

### Flow

Input

  |

Function Processing

   |

Return Result

    |

Store/ Reuse Result

```python
# another example:

def add(a, b):
    return a + b
```

```python
result = add(5, 10)
print("Result:", result) #because returned values can be returned.
```

```python
x = add(2,3)
y = add(5,7)  #without return this becomes impossible

print(x * y)
```

### scope

#### Variable Scope: Understanding Local and Global Variables

Core concept: Variables live in refions.

Not every variable exists everywhere.

```python
# Local Scope

def test():
    x = 10   # this works because x is defined within the function
```

```python
# Global Scope:

hospital = "Apollo"  #Accessible globally
```

```python
# Example:

def room():
    patient = "John"  # this is local to the function
    print(patient)
```

```python
# Outside:

try:
    print(patient)  # This will cause an error because 'patient' is not defined in the global scope
except NameError as e:
    print("Expected error:", e)
```

### Why Scope matters?

Prevent accidental modification.

Important in:

* ML systems
* hospital software
* APIs
* pipelines

### LEGB Rule

python searches for:

* Local
* Enclosing
* Global
* Built-in

```python
# another example:

x = 100 # this is a global variable and can be accessed anywhere in the code, including inside functions.

def test():
    x = 50  # this is a local variable that shadows the global variable x
    print(x)

test()
print(x) # this will print 100, the global variable, because the local variable x is only accessible within the function test()
```

### lambda

#### Lambda functions: Small Anonymous functions for Fast Data Operations

Key concept: Timny one-line function.

```python
# Normally:

def square(x):
    return x * x
```

```python
# Lambda function:

lambda x: x * x   #syntax: lambda parameters: expression
```

### Why Use Lambda

Useful for:

* sorting
* filtering
* transformations
* pamdas operations

Especially in data science.



```python
# Healthcare example: Sort patients by age:

patients = [
    ("John", 30), 
    ("Alice", 25), 
    ("Bob", 35)
]

patients.sort(key = lambda x: x[1])
```

```python
print(patients)  #sorted by age, ascending order
```

```python
# Using map()
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))
```

```python
print(numbers)  # original list

print(squared_numbers)  # squared list
```

```python
# Using filter()

bp = [120, 130, 140, 150, 160]
high_bp = list(filter(lambda x: x > 140, bp))
```

```python
print(high_bp)  # [150, 160]
```

### recursion

#### Recursion in Python: Functions That Call Themselves

Core Idea: A recursive function calls itself.

```python
# Simple example:

def countdown(n):
    if n == 0:
        print("Blast off!")
    else:
        print(n)
        countdown(n - 1)
```

### Why Recursion Matters

Used in:

* tree traversal
* file systems
* AI search
* NLP parsimg
* divide-and conquer algorithms

```python
# Factorial example:

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
```

## sets
### Title: Python Sets: Unique Data Handling for Clinical Intelligence

Set = Unordered collection of Unique elements.

```python
diagnosis_codes = {"I10", "E11", "J45"} #Why useful? Healthcare data often has duplicates.
```

```python
#Operations: add()
diagnosis_codes.add("C34")
```

```python
#remove()
diagnosis_codes.remove("I10")
```

```python
#discard()
diagnosis_codes = {"P001","I10", "E11", "J45"} #safer version
diagnosis_codes.discard("J45")
print(diagnosis_codes)
```

```python
#union()
a = {1,2,3}
b = {3,4,5}
result = a.union(b)
print(result)  #3 appears in both, only one in reult (sets automatically remove duplicates)
```

```python
#Multiple sets
lab_codes = {"CBC","HbA1c"}
diagnosis_codes = {"I10","E11"}
medication_codes = {"Metformin","Aspirin"}

everything = lab_codes.union(diagnosis_codes,medication_codes)

print(everything)
```

```python
#Union: shortcut(|)
a = {1,2,3}
b = {4,5,6}
print(a|b)
```

```python
#intersection: Common elements between sets
a = {1,2,3,4,5}
b = {3,4,6,7,5}
result = a.intersection(b)

print(result)
```

```python
# Health Example:
visit1 = {"Metformin", "Aspirin"}
visit2 = {"Metformin", "Atorvastatin"}

common = visit1.intersection(visit2)

print(common)
```

### Title:Python Dictionaries: Key-Value Modeling for Electronic Health Records

Dictionary = key-value storage.

```python
patient = {
    "id": "P001",
    "name": "Asha",
    "age": "43"
}
```

### Healthcare data is naturally structured

* age

```python
#Access:
patient["name"]
```

```python
# another method:
patient.get("name")
```

```python
#keys(): Returns all keys in the dictionary.
print(patient.keys()) #only field names
```

```python
# values(): Return all values
print(patient.values()) #Returns all values
```

```python
#items(): Returns key-value pairs.
print(patient.items())
```

```python
#update(): Adds new key-value pairs OR updates existing ones.
patient.update({"blood_group": "O+"})
print(patient)
```

```python
#Update existing:
patient.update({"age":25})
print(patient)
```

```python
#pop(): removes a specific key
patient.pop("id")
```

```python
#popitem(): Removes last inserted item
patient.popitem()
```

```python
#clear(): Deletes everything
patient.clear()
print(patient)
```

### Comprehensions

#### Efficient Clinical Data Transformation

Concept = Compact loops

```python
# Instead of

squares = []

for x in range(10):
    squares.append(x*x)
    
print(squares)
```

```python
#Use

squares = [x*x for x in range(10)]
print(squares)
```
